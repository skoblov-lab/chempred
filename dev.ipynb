```python
def exec_geniass(text: str) -> list:
    with open('geniass/data/tmp_in.txt', 'w') as f_in:
        f_in.write(text)

    p = subprocess.Popen(['./geniass', 'data/tmp_in.txt', 'data/tmp_out.txt'],
                         cwd='geniass/', stderr=subprocess.PIPE,
                         stdout=subprocess.PIPE)
    p.wait()
    out, err = p.communicate()

    with open('geniass/data/tmp_out.txt', 'r') as f_out:
        tmp_out = f_out.read()

    return tmp_out.split('\n')

def get_start_finish(text: str) -> pd.DataFrame:
    exec_geniass(text)
    
    p = subprocess.Popen(['ruby', 'sentence2standOff.rb', 'data/tmp_in.txt',
                          'data/tmp_out.txt', 'data/tmp_off.txt'], cwd='geniass/',
                         stderr=subprocess.PIPE, stdout=subprocess.PIPE)
    p.wait()
    out, err = p.communicate()
    out_decoded = out.decode()
    
    return get_df(out_decoded)
```
